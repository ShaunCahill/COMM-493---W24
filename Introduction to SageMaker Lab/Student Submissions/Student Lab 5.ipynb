```python
# 4.1 Drop the target variable
y = data.loc[:,"quality"].values
x = data.drop(['quality'], axis = 1)
```

```python
# Split the data into testing and training
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.33, random_state = 123)
```

```python
# 4.3 Setup the Sklearn Linear Regression

# Create linear regression object
from sklearn import linear_model
logreg = linear_model.LogisticRegression(max_iter=150)
```

```python
# 4.4 Train the model using the training dataset

# Train the model using the training sets
logreg.fit(x_train,y_train)
```

```python
# 4.5 Test the model using the testing dataset

# Make predictions using the testing set
predicted = logreg.predict(x_test)
```

```python
# 4.6 Comment on the acuracy of the model

# Show the accuracy of the model
print("Test accuracy: {} ".format(logreg.score(x_test, y_test)))
```

```python
# 4.7 Create a confusion matrix

# Create a confusion Matrix to identify TP, FP, FN, TN
from sklearn.metrics import confusion_matrix
cf_matrix = confusion_matrix(y_test,predicted)
cf_matrix
```

```python
# Create a heatmap of the Confusion Matrix
sns.heatmap(cf_matrix/np.sum(cf_matrix), annot=True, 
            fmt='.2%', cmap='Blues')
plt.show()
```

```python
# 4.8 Make a prediction

# Create your own DataFrame and run specific predictions
column_names=["volatile acidity", "total sulfur dioxide", "density", "chlorides", "pH", "free sulfur dioxide", "residual sugar", "fixed acidity", "citric acid", "sulphates", "alcohol"]
test_prediction_array = [ [9, 2, 5, 12, 121, 51, 4, 3, 48, 60, 8],
                          [8, 1, 13, 2, 159, 21, 3, 6, 54, 100, 7] ]
test_prediction_df = pd.DataFrame(test_prediction_array, columns=column_names)

test_prediction_df
```
